```
"""
In this script we are going to load the data from the CSV's to the staging tables.
"""
from scripts.helper import *
from pyspark.sql.functions import col,lit

PROCESS_DATE = '2023-07-02' # get_process_date() # date of data processed in YYYY-MM-DD format
```

```
# Load bi_corp_staging.calendar

df_calendar = spark_read_csv(spark, 'calendar', PROCESS_DATE, CSV_READER_OPTIONS)
df_calendar.printSchema()
df_calendar.show()
spark_write_df(df_calendar, STAGING_SCHEMA, 'calendar', STAGING)

%ls -Ralh /dbfs/FileStore/tables/bi_corp/staging/calendar
```

```
# Load bi_corp_staging.listings

df_listings = spark_read_csv(
    spark,
    'listings',
    PROCESS_DATE,
    {
        **CSV_READER_OPTIONS,
        "multiLine": True,
        "quote": '"',
        "escape": '"'
    }
)
#df_listings.printSchema()
df_listings.show()
spark_write_df(df_listings, STAGING_SCHEMA, 'listings', STAGING)

%ls -Ralh /dbfs/FileStore/tables/bi_corp/staging/listings
```
